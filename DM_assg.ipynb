```
#Processing indicators
#removing '..' where the values are missing
import csv
with open('Data.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data = list(spamreader)
    data = np.asarray(data)
    dt = pd.DataFrame(data=data[1:-1556], columns=data[0])
    dt.drop(columns=['Country Code', 'Series Name'], axis=1, inplace=True)
    dt.columns = ['Country', 'Series', '2013', '2014', '2015', '2016','2017','2018']
    dt.loc[dt['2015'] == '..', '2015'] = {}
    dt.loc[dt['2016'] == '..', '2016'] = {}
    dt.loc[dt['2017'] == '..', '2017'] = {}
    dt['2015'] = dt['2015'].astype(float)
    dt['2016'] = dt['2016'].astype(float)
    dt['2017'] = dt['2017'].astype(float)
    dt_train= dt[['Country', 'Series', '2015', '2016']]
    dt_test=dt[['Country','Series','2017']]
```

```
print (dt_train)
dt_train.to_csv('dt_train.csv', sep=',', encoding='utf-8')
```

```
print(dt_test)
dt_test.to_csv('dt_test.csv',sep=',',encoding='utf-8')
```

```
#Removing missing countries
countries = dt['Country'].unique()
dt_merge.drop(dt_merge.loc[~dt_merge['Country'].isin(countries)].index, inplace=True)
dt_merge.reset_index(inplace=True, drop=True)
dt_merge
```

```
dt_17.drop(dt_17.loc[~dt_17['Country'].isin(countries)].index, inplace=True)
dt_17.reset_index(inplace=True, drop=True)
dt_17
```

```
#Combining the datasets of the preprocessed 2015 and 2016 data with the World bank dataset containing the indicators
countries = dt_merge['Country'].unique()
for c in countries:
    indices = dt_train.loc[dt_train['Country'] == c].index
    if(indices.size == 0):
        continue
    
    c_ind = dt_merge.loc[dt_merge['Country'] == c].index
    for i in indices:
        dt_merge.at[c_ind[0], 'Indicator '+str(i%33+1)] = dt.loc[i]['2015']
        #dt_merge.at[c_ind[1], 'Indicator '+str(i%33+1)] = dt.loc[i]['2016']
        #dt_17.at[c_ind[2], 'Indicator '+str(i%33+1)] = dt.loc[i]['2017']
dt_merge.to_csv('dt_train1.csv', sep=',', encoding='utf-8')
#dt_17.to_csv('dt_test1.csv',sep=',',encoding='utf-8')
dt_merge
```

```
#Combining the datasets of the preprocessed 2017 dataset with the World bank dataset containing the indicators
countries = dt_17['Country'].unique()
for c in countries:
    indices = dt_test.loc[dt_test['Country'] == c].index
    if(indices.size == 0):
        continue
    
    c_ind = dt_17.loc[dt_17['Country'] == c].index
    for i in indices:
        dt_17.at[c_ind[0], 'Indicator '+str(i%33+1)] = dt_test.loc[i]['2017']
dt_17.to_csv('dt_test1.csv',sep=',',encoding='utf-8')
dt_17
```
